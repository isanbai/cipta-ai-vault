# 📘 Linear Regression – Prediksi Umur vs Tinggi

Silakan isi atau jalankan kode di bawah.

```python
# Linear Regression contoh sederhana
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data: umur (X) vs tinggi badan (y)
X = np.array([[5], [6], [7], [8], [9]])
y = np.array([110, 115, 120, 130, 135])

# Buat model regresi
model = LinearRegression()
model.fit(X, y)

# Visualisasi
plt.scatter(X, y, label='Data')
plt.plot(X, model.predict(X), color='red', label='Regresi')
plt.xlabel('Umur')
plt.ylabel('Tinggi Badan')
plt.title('Prediksi Tinggi berdasarkan Umur')
plt.legend()
plt.grid(True)
plt.show()
```
